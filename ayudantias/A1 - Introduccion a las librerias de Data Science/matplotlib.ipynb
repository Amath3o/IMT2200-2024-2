matplotlib es una librería de Python diseñada para poder realizar gráficos. En base a una colección de datos modelados en arrays (o incluso otras estrucutras de datos), podemos modelar estos datos usando gráficas de distintos tipos: Gráficos de bigotes, de barras, etc. 

Esta librería funciona muy bien con Numpy o con librerías badadas en Numpy (Como es el caso de Pandas), por lo que para este cuaderno vamos a trabajar con el dataset de Pandas usado previamente. 

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data/pokemon.csv')
df 
```

```python
#Nos qudamos con un subconjunto del dataset
sub_df = df.iloc[0:10]
```

## Y ahora procedemos a realizar distintas gráficas con los datos

En general, cuando definamos un gráfico, idealmente (Ya que no es necesario) podemos definir una tupla fig y ax para poder agregarle características a nuestro gráfico según lo que deseemos, con el objetivo de poder 'customizarlo' a nuestro antojo. En general esta tupla se define al crear el objeto **Subplots** que no es más que 'la base' del gráfico (Podríamos entenderlo como una imagen vacía) 

**fig, ax = plt.subplots()** 

Donde, esta imagen vacía se compone de:

* Fig (Figura): controla aspectos generales del gráfico, como el tamaño, la resolución y la disposición de los subgráficos

* ax (ejes) : El objeto Axes es donde realmente se dibujan los datos, y puedes tener múltiples Axes en un solo Figure. Es el lugar donde se trazan los gráficos, las etiquetas, los ticks y otros elementos del gráfico.


Para aprender a utilizar esta librería, se recomienda crear un subplots y luego ir agregándole cosas en función de lo que a uno se le ocurra, investigando directamente desde la documentación de la librería o de Stackoverflow.

https://matplotlib.org

```python
# gráfico de barras

fig, ax = plt.subplots() #creamos la figura y los ejes
ax.set_title('Peso de los 10 primeros pokemons') #título del gráfico
ax.set_xlabel('Nombre') # etiquetas de los ejes
ax.set_ylabel('Peso (kg)')  #etiquetas de los ejes
plt.bar(sub_df['name'], sub_df['weight_kg']) #creamos el gráfico de barras
plt.xticks(rotation=45) #rotamos las etiquetas del eje x
plt.show()
                                                            
```

```python
# gráfico de barras horizontal

fig, ax = plt.subplots() #creamos la figura y los ejes
ax.set_title('Peso de los 10 primeros pokemons') #título del gráfico
ax.set_xlabel('Peso (kg)') # etiquetas de los ejes
ax.set_ylabel('Nombre')  #etiquetas de los ejes
plt.barh(sub_df['name'], sub_df['weight_kg']) #creamos el gráfico de barras
plt.show()
```

```python
# gráfico de bigotes

fig, ax = plt.subplots() #creamos la figura y los ejes
ax.set_title('Peso de los 10 primeros pokemons') #título del gráfico
ax.set_xlabel('Peso (kg)') # etiquetas de los ejes
ax.set_ylabel('Nombre')  #etiquetas de los ejes
plt.boxplot(sub_df['weight_kg']) #creamos el gráfico de bigotes
plt.xticks([1], ['Peso']) #etiquetas de los bigotes
plt.show()
```

```python
# gráfico de puntos (scatter plot)

fig, ax = plt.subplots() #creamos la figura y los ejes
ax.set_title('Peso vs Altura de los 10 primeros pokemons') #título del gráfico
ax.set_xlabel('Peso (kg)') # etiquetas de los ejes
ax.set_ylabel('Altura (m)')  #etiquetas de los ejes
plt.scatter(sub_df['weight_kg'], sub_df['height_m']) #creamos el gráfico de puntos
plt.show()
```

```python
# Incluso podemos agregar multiples graficos en una misma figura y ejes

# tomamos un subconjunto del dataset
sub_df1 = df.iloc[0:10]
sub_df2 = df.iloc[10:20]

fig, ax = plt.subplots() #creamos la figura y los ejes
ax.set_title('Peso de los 10 primeros pokemons') #título del gráfico
ax.set_xlabel('Peso (kg)') # etiquetas de los ejes
ax.set_ylabel('Nombre')  #etiquetas de los ejes
plt.barh(sub_df1['name'], sub_df1['weight_kg'], color='blue') #creamos el gráfico de barras
plt.barh(sub_df2['name'], sub_df2['weight_kg'], color='red') #creamos el gráfico de barras
plt.show()

```

```python
# O separarlos en diferentes figuras ploteadas en una misma celda

fig, ax = plt.subplots() #creamos la figura y los ejes
ax.set_title('Peso de los 10 primeros pokemons') #título del gráfico
ax.set_xlabel('Peso (kg)') # etiquetas de los ejes
ax.set_ylabel('Nombre')  #etiquetas de los ejes
#fijamos un tamaño para la figura
fig.set_size_inches(10, 5)


plt.barh(sub_df1['name'], sub_df1['weight_kg'], color='blue') #creamos el gráfico de barras
plt.show()

fig, ax = plt.subplots() #creamos la figura y los ejes
ax.set_title('Peso de los 10 primeros pokemons') #título del gráfico
ax.set_xlabel('Peso (kg)') # etiquetas de los ejes
ax.set_ylabel('Nombre')  #etiquetas de los ejes
plt.barh(sub_df2['name'], sub_df2['weight_kg'], color='red') #creamos el gráfico de barras
plt.show()

```
