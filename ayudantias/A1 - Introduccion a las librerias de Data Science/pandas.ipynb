## 5. Filtrar: Obtener una fila (O conjuntos de fila) para un determinado Pokemon 

Pandas nos permite incorporar condiciones l√≥gicas para poder filtrar los datos seg√∫n lo que nos interese buscar

**Encontrar un pokemon tal que en su columna name se llame 'Pikachu'**

```python
df[df['name'] == 'Pikachu']
```

**Encontrar todos los Pokemon Legendarios**

```python
df[df['is_legendary'] == 1]
```

**Condiciones mas complejas como encontrar un pokemon tal que sea tipo fuego, acero y su ataque especial sea mayor a 100**

```python
fire_steel = df[(df['type1'] == 'fire') & (df['type2'] == 'steel') & (df['sp_attack'] > 100) | (df['type1'] == 'steel') & (df['type2'] == 'fire') & (df['sp_attack'] > 100)]
fire_steel
```

```python
fire_steel['name']
```
