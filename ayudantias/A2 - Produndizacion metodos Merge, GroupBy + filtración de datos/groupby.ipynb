# Profundización en Métodos de Pandas `groupby`

Pandas es una poderosa biblioteca de Python para el análisis de datos. A continuación, exploraremos en detalle algunos de los métodos más importantes en este caso: `groupby`.

## 1. `groupby`

```python
df_filtrado_region.head(10)
```

Para realizar `groupby` tenemos que identificar nuestras `key`, en este contexto de datos puede ser tanto la columna de `región` o `Código Región`, dado que son datos que se repiten dentro de nuestro dataframe

```python
df_covid_groupby_region = df_filtrado_region.groupby('Region')
df_covid_groupby_region
```

```python
df_covid_groupby_codigo_region = df_filtrado_region.groupby('Codigo region')
df_covid_groupby_codigo_region
```

Al ser un objeto, podemos instanciar diferentes métodos que existen en el, les mostrare algunos 

#### `sum()`, `max()` , `min()`, `median()` , `mean()`

```python
df_covid_groupby_region.sum()
```

Básicamente deja en desorden nuestro dataframe, dado que suma todas las columnas al no especificar cuales necesitamos sumar realmente 

```python
columnas_groupby = df_filtrado_region.columns[5:-1]  #especificamos las columnas que queremos sumar 


df_covid_sum = df_covid_groupby_region[columnas_groupby].sum(numeric_only=True)
df_covid_sum
```

`min()`

```python

df_covid_min = df_covid_groupby_region[columnas_groupby].min(numeric_only=True)
df_covid_min
```

`median()`

```python
df_covid_median = df_covid_groupby_region[columnas_groupby].median(numeric_only=True)
df_covid_median
```

`mean()`

```python
df_covid_mean = df_covid_groupby_region[columnas_groupby].mean(numeric_only=True)
df_covid_mean
```

#### `aggregate()`

```python
df_2020_groupby_aggregate = df_filtrado_region.groupby('Region')[columnas_groupby].aggregate(['min', np.median,  max])
df_2020_groupby_aggregate
```

Si queremos acceder a los datos de una fecha en específico nos dara un bloque

```python
dia_09_01_2023 = df_2020_groupby_aggregate['09-01-2023']
dia_09_01_2023
```

#### `transform()`

```python
columnas_adicionales = ['Region', 'Comuna']
df_2020_transform = df_filtrado_region.groupby('Region')[columnas].transform(lambda x: x)

df_2020_transform[columnas_adicionales] = df_filtrado_region[columnas_adicionales]

df_2020_transform.head()
```

### `apply()`

En resumen es para aplicar funciones dentro de nuestro objeto `groupby`

```python
#por x razón tenemos que sumar 1  a la cantidad de contagiados del día 30-03-2020 a todas las comunas de las regiones 

def sumar_1(data):

    data['30-03-2020'] = data['30-03-2020'] + 1


    return data.head()
    
```

```python
columnas_groupby = df_filtrado_region.columns[5:6]  #especificamos las columnas que queremos sumar 


df_covid_groupby_region[columnas_groupby]

display(df_filtrado_region[columnas_groupby].head(10) , df_covid_groupby_region[columnas_groupby].apply(sumar_1))

```
